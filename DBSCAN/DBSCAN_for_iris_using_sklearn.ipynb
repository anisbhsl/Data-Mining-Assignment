```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data=pd.read_csv('iris.data')
```

```python
##instantiate DBSCAN model
X=data.drop(['Iris-setosa'],axis=1,).values
db=DBSCAN(eps=0.5, min_samples=3, metric='euclidean')
db.fit(X)
```

```python
labels=db.labels_
```

```python
set(labels)
```

```python
silhouette_score(X,labels)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1,2,1)
iris_types = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
for iris_type in iris_types:
    plt.scatter(
        data['5.1'][data['Iris-setosa'] == iris_type],
        data['3.5'][data['Iris-setosa'] == iris_type],
        label=iris_type,s=200,cmap='viridis'
    )
plt.legend()
plt.title('Original data cluster')
plt.xlabel('sepal length')
plt.ylabel('sepal.width')




plt.subplot(1,2,2)
plt.scatter(X[:,0], X[:,1], alpha=0.2,
            s=200*X[3], c=labels, cmap='viridis')
plt.title('DBSCAN for iris SKLEARN')
plt.xlabel('sepal length')
plt.ylabel('sepal.width')
```
